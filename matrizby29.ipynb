```python
import numpy as np
import time
import matplotlib.pyplot as plt

# Matriz de distâncias 'bays29'
distance_matrix = np.array([
    [0, 107, 241, 190, 124, 80, 316, 76, 152, 157, 283, 133, 113, 297, 228, 129, 348, 276, 188, 150, 65, 341, 184, 67, 221, 169, 108, 45, 167],
    [107, 0, 148, 137, 88, 127, 336, 183, 134, 95, 254, 180, 101, 234, 175, 176, 265, 199, 182, 67, 42, 278, 271, 146, 251, 105, 191, 139, 79],
    [241, 148, 0, 374, 171, 259, 509, 317, 217, 232, 491, 312, 280, 391, 412, 349, 422, 356, 355, 204, 182, 435, 417, 292, 424, 116, 337, 273, 77],
    [190, 137, 374, 0, 202, 234, 222, 192, 248, 42, 117, 287, 79, 107, 38, 121, 152, 86, 68, 70, 137, 151, 239, 135, 137, 242, 165, 228, 302],
    [124, 88, 171, 202, 0, 61, 392, 202, 46, 160, 319, 112, 163, 322, 240, 232, 372, 300, 210, 21, 83, 385, 298, 82, 325, 55, 176, 117, 93],
    [80, 127, 259, 234, 61, 0, 386, 136, 117, 192, 317, 35, 127, 351, 268, 213, 366, 294, 204, 42, 96, 378, 291, 44, 338, 114, 157, 98, 181],
    [316, 336, 509, 222, 392, 386, 0, 380, 438, 195, 86, 421, 268, 9, 184, 159, 23, 51, 142, 372, 316, 37, 127, 338, 16, 393, 235, 366, 429],
    [76, 183, 317, 192, 202, 136, 380, 0, 217, 152, 229, 166, 47, 369, 300, 77, 405, 333, 243, 182, 141, 403, 216, 99, 392, 207, 28, 64, 239],
    [152, 134, 217, 248, 46, 117, 438, 217, 0, 206, 365, 152, 202, 425, 343, 269, 417, 345, 255, 66, 128, 430, 343, 162, 440, 101, 190, 152, 141],
    [157, 95, 232, 42, 160, 192, 195, 152, 206, 0, 115, 245, 74, 184, 102, 84, 217, 145, 55, 139, 96, 216, 204, 148, 208, 230, 127, 192, 256],
    [283, 254, 491, 117, 319, 317, 86, 229, 365, 115, 0, 353, 200, 75, 79, 143, 109, 37, 85, 299, 243, 122, 58, 270, 95, 376, 205, 237, 411],
    [133, 180, 312, 287, 112, 35, 421, 166, 152, 245, 353, 0, 162, 410, 328, 206, 401, 329, 239, 93, 147, 412, 325, 78, 405, 167, 137, 120, 234],
    [113, 101, 280, 79, 163, 127, 268, 47, 202, 74, 200, 162, 0, 257, 188, 30, 291, 219, 129, 142, 96, 305, 128, 81, 280, 203, 18, 70, 202],
    [297, 234, 391, 107, 322, 351, 9, 369, 425, 184, 75, 410, 257, 0, 169, 227, 32, 60, 151, 303, 247, 46, 127, 304, 14, 377, 240, 328, 313],
    [228, 175, 412, 38, 240, 268, 184, 300, 343, 102, 79, 328, 188, 169, 0, 158, 207, 135, 47, 220, 165, 219, 137, 224, 196, 275, 198, 262, 335],
    [129, 176, 349, 121, 232, 213, 159, 77, 269, 84, 143, 206, 30, 227, 158, 0, 182, 110, 20, 211, 135, 192, 85, 169, 181, 266, 48, 119, 271],
    [348, 265, 422, 152, 372, 366, 23, 405, 417, 217, 109, 401, 291, 32, 207, 182, 0, 74, 164, 353, 297, 59, 137, 319, 7, 306, 243, 302, 344],
    [276, 199, 356, 86, 300, 294, 51, 333, 345, 145, 37, 329, 219, 60, 135, 110, 74, 0, 92, 280, 224, 89, 11, 250, 65, 241, 170, 202, 278],
    [188, 182, 355, 68, 210, 204, 142, 243, 255, 55, 85, 239, 129, 151, 47, 20, 164, 92, 0, 190, 134, 175, 81, 160, 156, 262, 95, 166, 277],
    [150, 67, 204, 70, 21, 42, 372, 182, 66, 139, 299, 93, 142, 303, 220, 211, 353, 280, 190, 0, 62, 365, 278, 62, 372, 39, 157, 97, 126],
    [65, 42, 182, 137, 83, 96, 316, 141, 128, 96, 243, 147, 96, 247, 165, 135, 297, 224, 134, 62, 0, 278, 212, 87, 316, 66, 119, 77, 105],
    [341, 278, 435, 151, 385, 378, 37, 403, 430, 216, 122, 412, 305, 46, 219, 192, 59, 89, 175, 365, 278, 0, 127, 334, 24, 319, 238, 357, 357],
    [184, 271, 417, 239, 298, 291, 127, 216, 343, 204, 58, 325, 128, 127, 137, 85, 137, 11, 81, 278, 212, 127, 0, 238, 141, 302, 131, 164, 339],
    [67, 146, 292, 135, 82, 44, 338, 99, 162, 148, 270, 78, 81, 304, 224, 169, 319, 250, 160, 62, 87, 334, 238, 0, 348, 139, 78, 22, 215],
    [221, 251, 424, 137, 325, 338, 16, 392, 440, 208, 95, 405, 280, 14, 196, 181, 7, 65, 156, 372, 316, 24, 141, 348, 0, 308, 240, 327, 346],
    [169, 105, 116, 242, 55, 114, 393, 207, 101, 230, 376, 167, 203, 377, 275, 266, 306, 241, 262, 39, 66, 319, 302, 139, 308, 0, 191, 132, 38],
    [108, 191, 337, 165, 176, 157, 235, 28, 190, 127, 205, 137, 18, 240, 198, 48, 243, 170, 95, 157, 119, 238, 131, 78, 240, 191, 0, 56, 259],
    [45, 139, 273, 228, 117, 98, 366, 64, 152, 192, 237, 120, 70, 328, 262, 119, 302, 202, 166, 97, 77, 357, 164, 22, 327, 132, 56, 0, 196],
    [167, 79, 77, 302, 93, 181, 429, 239, 141, 256, 411, 234, 202, 313, 335, 271, 344, 278, 277, 126, 105, 357, 339, 215, 346, 38, 259, 196, 0]
])


# Função adaptada de distância
def distance(city1_idx, city2_idx):
    return distance_matrix[city1_idx][city2_idx]

def fitness(solution):
    return sum(distance(solution[i], solution[i+1]) for i in range(len(solution)-1)) + distance(solution[-1], solution[0])

def initialize_bats(num_bats=20):
    return [np.random.permutation(29) for _ in range(num_bats)]

def update_bat(bat, best_bat, alpha, gamma):
    new_bat = np.copy(bat)
    if np.random.rand() > gamma:
        i, j = np.random.choice(29, 2, replace=False)
        new_bat[i], new_bat[j] = new_bat[j], new_bat[i]
    else:
        indices = np.random.choice(29, size=int(alpha*29), replace=False)
        for idx in indices:
            new_bat[idx] = best_bat[idx]
    return new_bat

def bat_algorithm(num_bats=20, max_iter=1000, alpha=0.5, gamma=0.5):
    bats = initialize_bats(num_bats)
    best_bat = min(bats, key=fitness)

    for iteration in range(max_iter):
        for i in range(num_bats):
            if np.random.rand() < (1 - fitness(bats[i])/fitness(best_bat)):
                bats[i] = update_bat(bats[i], best_bat, alpha, gamma)
            if fitness(bats[i]) < fitness(best_bat):
                best_bat = np.copy(bats[i])
    return best_bat, fitness(best_bat)

def plot_cities_and_path(solution):
    # Aqui, nós não temos as coordenadas reais das cidades, então vamos criar um plot com base na ordem das cidades na solução
    plt.plot(np.arange(29), solution, 'o-')
    plt.ylabel('Índice da cidade')
    plt.xlabel('Ordem na solução')
    plt.title('Ordem das cidades na solução')
    plt.show()

# Imprimir matriz inicial
print("Matriz de Distâncias Inicial:")
print(distance_matrix)

# Inicia o cronômetro
start_time = time.time()

solution, distance_travelled = bat_algorithm()

# Calcula o tempo gasto
elapsed_time = time.time() - start_time

print("\nMelhor solução:", solution)
print("Distância percorrida:", distance_travelled)
print(f"Tempo de execução: {elapsed_time:.4f} segundos")

plot_cities_and_path(solution)
```
